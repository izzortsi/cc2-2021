```python
import numpy as np
from solver import Solver
```

```python

f = lambda t, y: -5 * y
y0 = 1
t0 = 0
tf = 2
h = [0.1, 0.05, 0.025,0.0125,0.00625]
analytical_solution = lambda t, y_0: y_0*np.exp(-5 * t)
```

```python
solver = Solver(method = "Forward Euler", analytical_solution = analytical_solution)
solver.multiple_plots(solver.forward_euler,h, f, y0, t0, tf)
```

```python
solver.method = "Backward Euler with Newton's method"
df = lambda t, y: -5
solver.multiple_plots(solver.backward_euler_newton, h, f, df, y0, t0, tf)
```

```python
solver.method = "Crank-Nicolson with Newton's method"
solver.multiple_plots(solver.crank_nicolson_newton, h, f, df, y0, t0, tf)
```

```python
solver.method = "Heun's method"
solver.multiple_plots(solver.heun_method, h, f, y0, t0, tf)
```
