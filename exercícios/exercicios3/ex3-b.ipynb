```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
e = np.linspace(0.1, 4.0, 10)
h = 0.01
t0 = 0.0
tf = 10.0
xs_0 = np.array([1.0, 0.0])
f = lambda t, x, e = 0.1: np.array([x[1], -e * (x[0]**2 - 1) * x[1] - x[0]])
# Number of steps
n_steps = int(np.ceil((tf - t0) / h))

# Time points
ts = np.linspace(t0, tf, n_steps + 1)

# Solution arrays
xs = np.zeros((n_steps + 1, 2))
xs[0,:] = xs_0

```

```python

# Forward Euler method
def forward_euler(ts, xs, f, h):
    for i in range(n_steps):
        x = xs[i]
        xs[i+1] = x + h * f(ts[i], x, )
    return xs

fig, ax = plt.subplots(figsize=(6, 10))
for e_i in e:
    xs = forward_euler(ts, xs, lambda t, x: f(t, x, e=e_i), h)
    
    ax.plot(xs[:,0], xs[:,1], "k--", label=f"e={e_i}", linewidth=0.2)
    ax.set_xlabel('u(t)')
    ax.set_ylabel('v(t)')
    ax.set_title('Phase plot')
```
